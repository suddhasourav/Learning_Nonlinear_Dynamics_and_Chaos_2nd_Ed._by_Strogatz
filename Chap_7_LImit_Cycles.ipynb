```python
import scipy as sp
import numpy as np
from matplotlib import pyplot as plt
```

```python
# This is Example 7.1.1 in Strogatz, but put in a Cartesian coordinate system.
# I also played around with the nonlinear terms' coefficients a bit
x = np.arange(-7, 7,0.2)
y = np.arange(-5,5,0.1)

X, Y = np.meshgrid(x,y)

 = Y
 = -np.sin(X) - 0.75*Y

plt.rcParams['figure.figsize'] = [10, 10]
speed = np.sqrt(**2 + **2)
speed_normalized = speed/speed.max()
plt.streamplot(X, Y, , , density=1, linewidth = 1,  color = np.sqrt(speed_normalized),
               cmap = 'magma', broken_streamlines=False, arrowstyle = '-')
plt.axis('equal')
plt.xlabel("X")
plt.ylabel("Y")

```

```python
# My first Van der Pol oscillator
# The equation can be made by serially complicating a linear harmonic oscillator:
#  + x = 0
# Now add a nonlinear damping term proportional to x:
#  - (x - 1) + x = 0
# We make a system with only one differentiation operator:
#  = y
#  = (x - 1)y - x
R = 5
x = np.arange(-R,R,0.1)
y = np.arange(-R,R,0.1)

 = 0.1

X, Y = np.meshgrid(x,y)
 = Y
 = *(X**2 - 1)*Y - X

speed = np.sqrt(**2 + **2)
speed_normalized = speed/speed.max()
plt.streamplot(X, Y, , , density=1, linewidth = 1,  color = np.sqrt(speed_normalized),
               cmap = 'magma', broken_streamlines=False)
plt.axis('equal')
plt.xlabel("X")
plt.ylabel("Y")
```
